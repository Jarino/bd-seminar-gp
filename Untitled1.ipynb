# 2D symbolic regression

```python
from pycgp.benchmarks.symbolic2d import X_train, y_train, X_test, y_test, x0, x1, y_truth, PARAMS, EV_PARAMS
from sklearn.metrics import mean_squared_error
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
```

```python
ax = plt.figure().gca(projection='3d')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
surf = ax.plot_surface(x0, x1, y_truth, rstride=1, cstride=1, color='green', alpha=0.5)
plt.show()
```

# MLP regressor

```python
from sklearn.neural_network import MLPRegressor
est = MLPRegressor(max_iter=100000, solver='lbfgs', hidden_layer_sizes=545, random_state=0)
est.fit(X_train, y_train)
```

```python
y_pred = est.predict(np.c_[x0.ravel(), x1.ravel()]).reshape(x0.shape)
mlp_score = est.score(X_test, y_test)

ax = plt.figure().gca(projection='3d')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
surf = ax.plot_surface(x0, x1, y_truth, rstride=1, cstride=1, color='green', alpha=0.5)
surf2 = ax.plot_surface(x0, x1, y_pred, rstride=1, cstride=1, color='yellow', alpha=0.5)
plt.show()
print(mlp_score)
```

```python
mean_squared_error(est.predict(X_test), y_test)
```

# PyCGP regression

```python
from copy import deepcopy
import random
from pycgp.mutation import probabilistic_mutation
from pycgp.evolution import evolution

params = deepcopy(PARAMS)
params['n_cols'] = 42
ev_params = deepcopy(EV_PARAMS)
ev_params['max_evaluations'] = 10000
ev_params['mutation'] = probabilistic_mutation

random.seed(0)

result = evolution(params, ev_params, X_train, y_train)

individual = result['final'][0]
```

```python
mean_squared_error(individual.execute(X_test), y_test)
```
