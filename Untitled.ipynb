```python
from pycgp.benchmarks.symbolic import X, y, PARAMS, EV_PARAMS
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%time

rgr = MLPRegressor(max_iter=100000, solver='lbfgs', hidden_layer_sizes=100, random_state=1)
rgr.fit(X, y)
```

```python
y_pred = rgr.predict(X)
plt.plot(X, y, label='gtruth')
plt.plot(X, y_pred, label='MLP')
print(mean_squared_error(y_pred, y))
```

```python
%%time

from pycgp.evolution import evolution
from pycgp.mutation import probabilistic_mutation
from pycgp.gems import GemSM, MatchByActiveStrategy
from pycgp.benchmarks.symbolic import PARAMS, EV_PARAMS
import random

PARAMS['max_evaluations'] = 5000
PARAMS['n_cols'] = 50
PARAMS['n_rows'] = 10
EV_PARAMS['mutation'] = probabilistic_mutation
EV_PARAMS['gems'] = True
EV_PARAMS['gem_type'] = GemSM
EV_PARAMS['match_strategy'] = MatchByActiveStrategy
random.seed(0)

results = evolution(PARAMS, EV_PARAMS, X, y)
individual = results['final'][0]
```

```python
y_pred = individual.execute(X)
plt.plot(X, y, label='gtruth')
plt.plot(X, y_pred, label='MLP')
print(mean_squared_error(y_pred, y))
```

```python
from gplearn.functions import make_function

gp_learn_funset = ['add', 'sub', 'mul', 'div', 'sin', 'cos', 'log']
    
```

```python
est_gp = SymbolicRegressor(population_size=5000, metric='mse',
                           generations=20, stopping_criteria=0.01,
                           p_crossover=0.7, p_subtree_mutation=0.1,
                           p_hoist_mutation=0.05, p_point_mutation=0.1,
                           max_samples=0.9, verbose=1,
                           parsimony_coefficient=0.01, random_state=1, function_set=gp_learn_funset)

est_gp.fit(X,y)
```

```python
y_pred = est_gp.predict(X)
plt.plot(X, y, label='gtruth')
plt.plot(X, y_pred)
```
